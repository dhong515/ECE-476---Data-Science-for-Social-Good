```python
#The 'spanish-translated.csv' and 'english-organized.csv' files were combined into one file called: 'combined.csv'.

import xlrd
book = xlrd.open_workbook('combined.xlsx')
sheet = book.sheet_by_name('Sheet1')
corpus = [[sheet.cell_value(r, c) for c in range(sheet.ncols)] for r in range(sheet.nrows)]
corpus = [i[0] for i in corpus]
print("Sentence List: ", corpus)
print("Total Sentences: ", len(corpus))
```

**Bidirectional Encoder Representations from Transformers (BERT)** is a machine learning technique for natural language processing pre-training that was developed by Google in 2018

**Embedding** is a process where words or phrases of a document are mapped to vectors of real numbers called embeddings.

**Sentence-transformers** is a framework that provides an easy method to compute dense vector representations for sentences and paragraphs.

```python
#Uncomment this line out before running if this package has not been not downloaded yet. 
#!pip install sentence-transformers

from sentence_transformers import SentenceTransformer

model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')
doc_embedding = model.encode(corpus)
```

**Clustering** is a task of grouping a set of objects so that objects in the same group are similar to each other than the objects in other groups.

The **K-Means Clustering** algorithm aims to partition n observations into k clusters (the value of k needs to be set by the user) in which each observation belongs to the cluster with the nearest mean.

```python
from sklearn.cluster import KMeans

num_clusters = 4

clustering_model = KMeans(n_clusters=num_clusters) # Defining the k-means model
clustering_model.fit(doc_embedding) # Fitting the embedding with kmeans clustering.
cluster_assignment = clustering_model.labels_ # Getting the cluster id assigned to each sentence.
cluster_assignment_list = [x + 1 for x in cluster_assignment]

print("Cluster Assignments: " , cluster_assignment_list, "\n")

clustered_sentences = [[] for i in range(num_clusters)]

for sentence_id, cluster_id in enumerate(cluster_assignment):
    clustered_sentences[cluster_id].append(corpus[sentence_id])

for i, cluster in enumerate(clustered_sentences):
    print("Cluster ", i + 1)
    print(cluster)
    print("Total number of sentences in cluster: ", len(cluster), "\n")
```

**Sentiment analysis** is a natural language processing method used to determine the emotion expressed in the text. 

The **pipeline** class was utlized in order to analyze the comments through the default sentiment analysis model in Python. 

Each comment is labeled as being either “Negative” or “Positive”, and an associated score is provided for each. 

The scores represent the accuracy in predicting the labels, and so scores closer to one are more accurate.


```python
#Uncomment this line out before running if this package has not been not downloaded yet. 
#!pip install -q transformers

from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis", model = "distilbert-base-uncased-finetuned-sst-2-english")
negative_count = 0
positive_count = 0

for i in range (len(corpus)):
  print(corpus[i])
  sentiment_results = sentiment_pipeline(corpus[i])[0]
  print(sentiment_results, "\n")

  if sentiment_results['label'] == 'NEGATIVE':
    negative_count = negative_count + 1
  else:
    positive_count = positive_count + 1

print("Total POSITIVE count = ", positive_count)
print("Total NEGATIVE count = ", negative_count)
```

**Rake** (Rapid Automatic Keyword Extraction) is a natural language processing method of extracting keywords from individual documents. 

It can also be applied to new fields very easily and is very effective in dealing with multiple types of documents, especially 
text that requires specific grammatical conventions. 

Rake identifies key phrases in a text by analyzing the occurrence of a word 
and its compatibility with other words in the text (co-occurrence). 

An individual **word score** is calculated as the degree 
(number of times it appears + number of additional words it appears with) of a word divided by its frequency (number of times it appears).

```python
#Uncomment this line out before running if this package has not been not downloaded yet. 
#!pip install multi_rake

from multi_rake import Rake

rake = Rake()
for i in range (len(corpus)):
  keywords = rake.apply(corpus[i])
  print("Sentence: ", corpus[i])
  print("Keywords: ", keywords, "\n")
```
